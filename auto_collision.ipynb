```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Create database from insurance claim file
claims_file = "Resources/insurance_claims.csv"
claims_df = pd.read_csv(claims_file)
claims_df.head()
```

```python
claims_df["umbrella_limit"].replace(0,claims_df["umbrella_limit"].mean(),inplace=True)



```

```python
claims_df.head()
```

```python
#Create database from accident file
accident_file = "Resources/ACC_AUX.csv"
accident_df = pd.read_csv(accident_file)
accident_df.head()
```

```python
car_type = ["auto_make", "auto_model", "auto_year"]
car_type_df = claims_df[car_type].copy()

car_type_df = car_type_df.rename(columns={"auto_make": "Make",
                                         "auto_model": "Model",
                                         "auto_year": "Year"})

car_type_df.to_csv("Car_Type.csv", index=False)

car_type_df.head()
```

```python
driver = ["insured_sex", "age", "insured_education_level", "policy_state", "incident_state"]
driver_df = claims_df[driver].copy()

driver_df = driver_df.rename(columns={"insured_sex": "Sex",
                                     "age": "Age",
                                     "insured_education_level": "Education",
                                     "policy_state": "Policy_State",
                                     "incident_state": "Incident State"})

driver_df.to_csv("Driver.csv", index=False)

driver_df.head()                                     
```

```python
insurance_policy = ["policy_state", "policy_number", "policy_deductable", "policy_csl", "policy_bind_date", "policy_annual_premium", "umbrella_limit", "fraud_reported"]
insurance_policy_df = claims_df[insurance_policy].copy()

insurance_policy_df = insurance_policy_df.rename(columns={"policy_state": "Policy_State",
                                   "policy_number": "Policy_Number",
                                   "policy_deductable": "Deductable",
                                   "policy_csl": "CSL",
                                   "policy_bind_date": "Policy_Start_Date",
                                   "policy_annual_premium": "Policy_Premium",
                                   "umbrella_limit": "Umbrella_Limit",
                                   "fraud_reported": "Fraud_Reported"})

insurance_policy_df.to_csv("Incident_Policy.csv", index=False)

insurance_policy_df.head()
```

```python
incident_location = ["incident_state", "insured_zip", "incident_city", "incident_date", "incident_location"]
incident_location_df = claims_df[incident_location].copy()

incident_location_df = incident_location_df.rename(columns={"incident_state": "State",
                                                           "insured_zip": "Zipcode",
                                                           "incident_city": "City",
                                                           "incident_date": "Incident_Date",
                                                           "incident_location": "Incident_Location"})

incident_location_df.to_csv("Incident_Location.csv", index=False)

incident_location_df.head()
```

```python
date = ["incident_date", "policy_bind_date"]
date_df = claims_df[date].copy()

date_df = date_df.rename(columns={"incident_date": "Incident_Date",
                                 "policy_bind_date": "Policy_Start_Date"})

date_df.to_csv("Date.csv", index=False)

date_df.head()
```

```python
incident = ["bodily_injuries", "property_damage", "incident_hour_of_the_day", "number_of_vehicles_involved", "police_report_available","incident_date", "incident_type", "collision_type", "incident_state"]
incident_df = claims_df[incident].copy()

incident_df = incident_df.rename(columns={"bodily_injuries": "Bodily_Injury",
                                         "property_damage": "Property_damage",
                                         "incident_hour_of_the_day": "Incident_Hour_of_Day",
                                         "number_of_vehicles_involved": "Number_of_Vehicles",
                                         "police_report_available": "Police_Report",
                                         "incident_date": "Incident_Date",
                                         "incident_type": "Incident_Type",
                                         "collision_type": "Collission_Type",
                                         "incident_state": "Incident_State"})

incident_df.to_csv("Incident_Type.csv", index=False)

incident_df.head()
```

```python
insurance_claim = ["policy_number", "total_claim_amount", "injury_claim", "property_claim", "vehicle_claim"]
insurance_claim_df = claims_df[insurance_claim].copy()

insurance_claim_df = insurance_claim_df.rename(columns={"policy_number": "Policy_Number",
                                                       "total_claim_amount": "Total_Claim",
                                                       "injury_claim": "Injury_Claim",
                                                       "property_claim": "Property_Claim",
                                                       "vehicle_claim": "Vehicle_Claim"})

insurance_claim_df.to_csv("Insurance_Claim.csv", index=False)

insurance_claim_df.head()
```
